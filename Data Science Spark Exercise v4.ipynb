```python
import findspark
findspark.init()
from pyspark.sql import SparkSession
from pyspark.conf import SparkConf
spark=SparkSession.builder\
    .master("local[*]")\
    .appName("WordCount")\
    .getOrCreate()
sc=spark.sparkContext
```

```python
lines = sc.textFile(r"C:\Users\Daniel\Desktop\Data Science\shakespeares.txt")

#seperate words
words = lines.flatMap(lambda line: line.split(" "))

#count the words by how often they occur in the text
wordCounts = words.countByValue()

#Sort the wordCounts in descending order by their occurence
wordCountsSorted = dict(sorted(wordCounts.items(), key= lambda item: item[1], reverse = True))
wordCountsSorted
```

```python
print(list(wordCountsSorted.items())[24])
```
